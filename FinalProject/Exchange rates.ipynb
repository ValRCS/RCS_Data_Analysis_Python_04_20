```python
df = pd.read_html("https://www.eurex.lv/valutas-kursi/valutas-kursi/")
df[2]
```

```python
dfe = df[2].drop(columns=['Unnamed: 5','Valūta']).rename(columns={'Unnamed: 0':'Currency', 'Pirkšana': 'EUREX Pirkšana', 'Pārdošana':'EUREX Pārdošana', 'Nosaukums':'EUREX Nosaukums'})
dfe
```

```python
dff = pd.read_html('https://www.swedbank.lv/private/d2d/payments2/rates/currency')
dff[0]
```

```python
dfs = dff[0].rename(columns={'Unnamed: 0_level_0':'Currency1','Klients pārdod *':'SWED Pirkšana', 'Klients pērk *':'SWED Pārdošana'})
dfs
```

```python
dfs.columns = dfs.columns.droplevel(1)
dfs
```

```python
dfs = dfs.drop(index=[27,28])
dfs
```

```python
new = dfs["Currency1"].str.split('.\(', expand=True)
dfs["Currency"]= new[0]
dfs['Nosaukums']= new[1]
dfs.drop(columns =["Currency1"], inplace = True) 
dfs
```

```python
result = pd.merge(dfe,dfs, left_on='Currency', right_on='Currency', how='inner')
result
```

```python
results = result.drop(columns=["Nosaukums"])
results
```

```python
results = results.rename(columns={'EUREX Nosaukums':'Nosaukums'})
results
```

```python
results.dtypes
```

```python
results[['SWED Pirkšana','SWED Pārdošana','ECB kursi']] = results[['SWED Pirkšana','SWED Pārdošana','ECB kursi']].astype(float)
results.dtypes
```

```python
results[['Currency']] = results[['Currency']].astype(str)
results.dtypes
```
